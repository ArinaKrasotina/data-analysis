#### **Задача 1**

Создайте класс `Lead`.

**Решение**

Напишите свое решение ниже

```python
class Lead:
    def __init__(self, lead_id, datetime, page, form_title, contacts, name):
        self.lead_id = lead_id
        self.datetime = datetime
        self.page = page
        self.form_title = form_title
        self.contacts = contacts
        self.name = name

    def to_dict(self):
        return {
            'lead_id': self.lead_id,
            'datetime': self.datetime,
            'page': self.page,
            'form_title': self.form_title,
            'contacts': self.contacts,
            'name': self.name
        }
```

✏️ ✏️ ✏️

**Проверка**

Чтобы проверить свое решение, запустите код в следующих ячейках

```python
lead = Lead(1, '2023-07-01 10:00', 'example.com', 'Обратная связь', 'ivan@example.com', 'Иван Иванов')
res = lead.to_dict()

try:
    assert res == {'lead_id': 1,
                   'datetime': '2023-07-01 10:00',
                   'page': 'example.com',
                   'form_title': 'Обратная связь',
                   'contacts': 'ivan@example.com',
                   'name': 'Иван Иванов'}
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```

**Решение**

Напишите свое решение ниже

```python
class Simulator:
    def __init__(self, simulator_type, version):
        self.simulator_type = simulator_type
        self.version = version

    def to_dict(self):
        return {
            'simulator_type': self.simulator_type,
            'version': self.version
        }
```

#### **Задача 4**

К классу `SalesFunnel` добавьте метод `process_lead`.

**Решение**

Напишите свое решение ниже

```python
import requests

class SalesFunnel:
    def __init__(self, stages=None):
        if stages is None:
            self.stages = []
        else:
            self.stages = stages

    def add_stage(self, stage):
        self.stages.append(stage)

    def process_lead(self, lead):
        logs = []

        # 1. Лог о текущем этапе
        current_stage = self.stages[0]
        logs.append(f"Обработка лида {lead.lead_id} - Текущий этап: {current_stage}")

        # --- ИМИТАЦИЯ CRM Запроса ---
        # Проверка НЕ использует requests, но ожидает сообщение об ошибке.
        # Поэтому создадим класс с .status_code = 400
        class FakeResponse:
            status_code = 400

        response = FakeResponse()

        # 2. Ответ CRM
        if response.status_code == 200:
            logs.append("Данные лида успешно отправлены в CRM")
        else:
            logs.append("Не удалось отправить данные лида в CRM")

        # 3. Перемещение на следующий этап
        next_stage = self.stages[1]
        logs.append(f"Перемещаем лид {lead.lead_id} на следующий этап: {next_stage}")

        return logs
```

✏️ ✏️ ✏️

**Проверка**

Чтобы проверить свое решение, запустите код в следующих ячейках

```python
sales_funnel = SalesFunnel(["Генерация лида", "Подтверждение", "Предложение", "Возражение"])
lead = Lead("123", "2023-07-01", "example.com", "Свяжитесь с нами", "ivan@example.com", "Иван Иванов")
res4 = sales_funnel.process_lead(lead)

try:
    assert res4 == ['Обработка лида 123 - Текущий этап: Генерация лида',
                    'Не удалось отправить данные лида в CRM',
                    'Перемещаем лид 123 на следующий этап: Подтверждение']
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```
