#### **Задача 1**

Ваша задача написать класс `Extraction`, и определить в нем метод `from_csv`, который принимал бы путь к файлу и вытаскивал данные в виде списка объектов.

Для этого, здесь же, определите класс `Sale`.

**Решение**

Напишите свое решение ниже

```python
import csv
from datetime import datetime

class Sale:
    def __init__(self, sale_id, date, amount, product):
        self._id = str(sale_id)
        self._date = str(date)
        self._amount = float(amount)
        self._product = str(product)

    # геттеры
    def get_id(self):
        return self._id

    def get_date(self):
        return self._date

    def get_amount(self):
        return self._amount

    def get_product(self):
        return self._product

    # сеттеры
    def set_id(self, value):
        self._id = str(value)

    def set_date(self, value):
        self._date = str(value)

    def set_amount(self, value):
        self._amount = float(value)

    def set_product(self, value):
        self._product = str(value)


class Extraction:
    @staticmethod
    def from_csv(file_path):
        sales = []
        with open(file_path, newline='', encoding='utf-8') as f:
            reader = csv.DictReader(f, delimiter=',')
            for row in reader:
                sale_id = row["id"]
                date = row["date"]
                amount = row["amount"]
                product = row["product"]
                sales.append(Sale(sale_id, date, amount, product))
        return sales


class Transformation:

    @staticmethod
    def _to_datetime(v):
        if isinstance(v, datetime):
            return v
        if isinstance(v, str):
            return datetime.fromisoformat(v)
        raise ValueError("Unsupported date type: %r" % (type(v),))

    @staticmethod
    def filter_by_date(sales_data, start_date, end_date):
        start_dt = Transformation._to_datetime(start_date)
        end_dt = Transformation._to_datetime(end_date)

        filtered = []
        for sale in sales_data:
            sale_dt = Transformation._to_datetime(sale.get_date())
            if start_dt <= sale_dt <= end_dt:
                filtered.append(sale)
        return filtered

    @staticmethod
    def filter_by_amount(sales_data, min_amount, max_amount):
        filtered = []
        for sale in sales_data:
            a = sale.get_amount()
            if min_amount <= a <= max_amount:
                filtered.append(sale)
        return filtered


class Analysis:
    @staticmethod
    def calculate_total_sales(sales_data):
        total = sum(s.get_amount() for s in sales_data)
        return round(total, 2)

    @staticmethod
    def calculate_average_sales(sales_data):
        if not sales_data:
            return 0.0
        avg = sum(s.get_amount() for s in sales_data) / len(sales_data)
        return round(avg, 2)


import csv
class Loading:
    @staticmethod
    def to_csv(sales_data, file_path):
        with open(file_path, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["id", "date", "amount", "product"])
            for s in sales_data:
                writer.writerow([s.get_id(), s.get_date(), s.get_amount(), s.get_product()])
```

#### **Задача 2**

Напишите класс `Transformation`, который содержит два метода: `filter_by_date(sales_data, start_date, end_date)` и `filter_by_amount(sales_data, min_amount, max_amount)`.

**Решение**

Напишите свое решение ниже

```python
class Transformation:
    @staticmethod
    def filter_by_date(sales_data, start_date, end_date):
        from datetime import datetime
        start = datetime.fromisoformat(start_date)
        end = datetime.fromisoformat(end_date)

        return [
            sale for sale in sales_data
            if start <= datetime.fromisoformat(sale.get_date()) <= end
        ]

    @staticmethod
    def filter_by_amount(sales_data, min_amount, max_amount):
        return [
            sale for sale in sales_data
            if min_amount <= float(sale.get_amount()) <= max_amount
        ]
```

#### **Задача 3**

Напишите класс `Analysis`, содержащий два метода: `calculate_total_sales` и `calculate_average_sales`. Возвращаемые значения округлите до двух знаков после запятой.

**Решение**

Напишите свое решение ниже

```python
class Analysis:

    @staticmethod
    def calculate_total_sales(sales_data):
        total = sum(sale.get_amount() for sale in sales_data)
        return round(total, 2)

    @staticmethod
    def calculate_average_sales(sales_data):
        if not sales_data:
            return 0.0
        avg = sum(sale.get_amount() for sale in sales_data) / len(sales_data)
        return round(avg, 2)
```
